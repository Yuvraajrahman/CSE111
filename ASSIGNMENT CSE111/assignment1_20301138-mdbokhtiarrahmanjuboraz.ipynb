```python
#LIST_04_SOLUTION

def sort(list_a):
    length = len(list_a)-1
    sorted = False


    while not sorted:
        sorted = True
        for i in range(0, length):
            if list_a[i] > list_a[i+1]:
                sorted = False
                list_a[i], list_a[i+1] = list_a[i+1], list_a[i]
    return list_a


list1 = [1, 4, 2, 3]
list1 = sort(list1)

for count in range(len(list1)):
    yes_no = 0
    if count+1 == list1[count]:
        yes_no = 1
        print('UB Jumper')
    else:
        print('Not UB Jumper')
    break
```
