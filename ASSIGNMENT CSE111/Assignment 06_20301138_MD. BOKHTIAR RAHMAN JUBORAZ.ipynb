```python
#solution no.2


class Dates():
    def __init__(self, d1):
        self.d1 = d1

    @classmethod
    def toDashDate(cls, s1):
        d2 = s1.replace("/","-")

        obj = cls(d2)

        return obj

    def getDate(self):
        return self.d1


date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate()):
 print("Equal")
else:
 print("Unequal")

#subtask :
print("The conditional statement prints 'Equal' because we replaced all '/' to '-' from 'dateFromDB' object")
```

```python
#solution no.5

class Fruit:
    fruitCount = 0
    def __init__(self, name, count):
        self.name = name
        self. count = count
        Fruit.fruitCount += self.count

    @staticmethod
    def saySomethingGood():
        print("Fruits are good for health")

    @classmethod
    def resetcount(cls):
        Fruit.fruitCount = 0






apples = Fruit("Apple", 3);
pears = Fruit("Pear", 4);
print(apples.name, apples.count)
print(pears.name,pears.count)
print("Total number of fruits", Fruit.fruitCount)
Fruit.saySomethingGood()
Fruit.resetcount()
print("Total number of fruits", Fruit.fruitCount)

```

```python
#solution no.6
class Cat:
    Number_of_cats = 0

    def __init__(self, c, a ):
        self. c = c
        self. a = a
        Cat.Number_of_cats += 1

    @classmethod
    def no_parameter(cls):
        c = 'White'
        a = 'sitting'
        obj = cls(c, a)
        return obj

    @classmethod
    def first_parameter(cls,x):
        c = x
        a = 'sitting'
        obj = cls(c,a)
        return obj

    @classmethod
    def second_parameter(cls, x):
        c = 'White'
        a = x
        obj = cls(c, a)
        return obj

    def changeColor(self, x):
        self.c = x

    def printCat(self):
        print(f'{self.c} cat is {self.a}.')



print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#solution no.7
class Cylinder():
    radius = 5
    height = 18

    def __init__(self, radius, height):
        print(f'Default radius={Cylinder.radius} and  height={Cylinder.height}.')
        self.radius = radius
        self.height = height
        print(f'Updated: radius={self.radius} and height={self.height}.')
        Cylinder.radius, Cylinder.height = self.radius, self.height

    @staticmethod
    def volume(x, y):
        valume = 3.14159265 * x * x * y
        print('Volume:', valume)

    @staticmethod
    def area(r1, h1):
        pi = 3.14159265

        area = (2 * pi * r1 * r1) + (2 * pi * r1 * h1)
        print('Area:', area)
    @classmethod
    def swap(cls, x, y):
        radius = y
        height = x
        obj1 = cls(radius, height)
        return obj1

    @classmethod
    def changeFormat(cls, x):
        radius, height = x.split('-')
        obj2 = cls(int(radius),  int(height))

        return obj2


c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```
