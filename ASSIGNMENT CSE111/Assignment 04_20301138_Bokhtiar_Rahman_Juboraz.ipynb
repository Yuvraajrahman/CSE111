```python
#SOLUTION OF QUESTION NO. 02
class Customer():
    def __init__(self,  name):

        self.name = name



    def greet(self, name=None):

        if name == None:
            print('Hello!')
        else:
            print(f'Hello {self.name}!')

    def purchase(self, *val):

        num_items = len(val)

        print(f'{self.name} have purchased {num_items} items(s):')
        for i in val:
            print(i)
#============================================================



customer_1 = Customer("Sam")
customer_1.greet()

customer_1.purchase("chips", "chocolate", "orange  juice")
print("-----------------------")


customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")

```

```python
#SOLUTION OF QUESTION NO. 05

class Vehicle():
    def __init__(self, x=0, y=0):
        
        self.x = x
        self.y = y

    def print_position(self):
        
        print(f'({self.x},{self.y})')


    def moveUp(self):
        self.y += 1



    def moveDown(self):
        self.y -= 1


    def moveLeft(self):
        self.x -= 1

    def moveRight(self):
        self.x += 1
#========================================

car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#SOLUTION OF QUESTION NO. 06
class Programmer():
    def __init__(self, name, language, experiece):

        self.name = name
        self.experience = experiece
        self.language = language


        print('Horray! A new programmer is born')



    def printDetails(self):
        print('Name:', self.name)
        print('Language:', self.language)
        print('Experience:', self.experience)


    def addExp(self, n1):
        print(f'Updating experience of {self.name}')
        self.experience += n1
#=======================================================

p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()


```

```python
#SOLUTION OF QUESTION NO. 07
class Student():
    def __init__(self, name, id, department = 'CSE'):
        self.name = name
        self.id = id 
        self.department = department
    
    def printDetails(self):
        print('Name:',self.name)
        print('ID:',self.id)
        print('Department:',self.department)
        print('Daily Effort:',self.hours,'hour(s)')
        s = ''
        if self.hours <= 2:
            s = 'Should give more effort!'
        elif self.hours <= 4:
            s = 'Keep up the good work!'
        else:
            s = 'Excellent! Now motivate others.'
        print(f'suggetion:{s}')

    def dailyEffort(self, hours):
        self.hours = hours
        

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
#SOLUTION OF QUESTION NO. 08
class Patient():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def add_Symptom(self, *sym):
        self.sym = sym
    
    def printPatientDetail(self):
        print('Name:',self.name)
        print('Age:',self.age)
        sym = ''
        for i in self.sym:
            sym += i + ',' 
        print('Symptoms:',sym)

p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
#SOLUTION OF QUESTION NO. 10
class Shinobi():
    def __init__(self, name, rank, point = 0):
        self.point = point
        self.name = name
        self.rank = rank





    def printInfo(self):
        print('Name:', self.name)
        print('Rank:', self.rank)
        print('Number of mission:', self.point)
        
        salary = 0
        if self.rank == 'Genin':
            salary = self.point * 50
        elif self.rank == 'Chunin':
            salary = self.point * 100
        else:
            salary = self.point * 500
        print('Salary:',salary)


    def calSalary(self,point):
        self.point = point

    def changeRank(self,rank):
        self.rank = rank

#=====================================================
naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```
